This just plots the data collected from keystone.

```python
import numpy as np
from util import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_data, optimal_train, test_data, optimal_test = load_keystone_data('trainErrors.csv', 'testErrors.csv')
```

```python
mean_train = np.mean(train_data, axis=0)
mean_test = np.mean(test_data, axis=0)
train_std = np.std(train_data)
test_std = np.std(test_data)

x = list(range(99))
#plt.errorbar(x, mean_train, yerr=train_std, label='Incremental')
fig, axs = plt.subplots(nrows=2, ncols=1, sharex=True)


axs[0].set_title('Train error')
axs[0].plot(mean_train, label='Incremental')
axs[0].plot(optimal_train, label='Optimal')
axs[0].legend()

#plt.errorbar(x, mean_test, yerr=test_std, label='Incremental')
axs[1].set_title('Test error')
axs[1].plot(mean_test, label='Incremental')
axs[1].plot(optimal_test, label='Optimal')
axs[1].set_xlabel('percent data')
axs[0].set_ylabel('percent error')
plt.show()
```

The incremental version seems to outperform the optimal version in training error, but this may be because the train error test is on the local batch, and in the optimal case this batch is the aggregate data up to the current iteration. So the optimal case could just be harder.

Also, this might cause the low train error in the beggining - it is easier to overfit to the small number of examples given more time to converge.
